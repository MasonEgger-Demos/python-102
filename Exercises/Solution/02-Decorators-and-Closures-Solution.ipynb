# Exercise #1 - Write a Debugging Decorator

Write a decorator that prints the name of the function, along with its arguments. The original function should then be executed. Afterwards the decorator should print the result to the screen and return the result.

**Hint**: You can get the name of a function by using `FUNCTION_NAME.__name__`

```python
def my_func():
  pass

my_func.__name__
```

**CHALLENGE**: Write the decorator to take _any_ number of positional and keyword arguments

````python
# Challenge Answer

def debug(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with arguments: {args}, {kwargs}")
        result = func(*args, **kwargs)
        print(f"Return value: {result}")
        return result
    return wrapper

# Example usage:
@debug
def add(a, b):
    return a + b

add(3, 5)
````

# Exercise #2 - Write a Decorator that Always Gives the Answer to the Previous Call

Write a silly decorator that makes a function return the result from the previous execution. You can have the function return whatever you want for the first execution, but every subsequent execution must return the result of the previous invokation.

**Hint**: Use closures and `nonlocal`.


````python
def previous(func):
    prev_answer = "undefined"

    def wrapper(a,b):
        nonlocal prev_answer
        result = prev_answer
        prev_answer = func(a,b)
        return result
    return wrapper

@previous
def add(a, b):
    return a + b

print(add(1, 2))
print(add(3, 4))
print(add(5, 6))
````
