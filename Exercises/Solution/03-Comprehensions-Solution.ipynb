# Exercise #1 - Write a Initial Creator

Write a list comprehension that creates the intials of a person based on their first and last name. A list is provided containing the names of people, their first name sepearate from the last name by a space.



```python
names = ["Mason Egger", "Carol Willing", "Lynn Root", "Pandy Knight"]
initials = [name.split()[0][0] + name.split()[1][0] for name in names]
print(initials)  # Output: ['ME', 'CW', 'LR', 'PK']
```

# Exercise #2 - Write a Vowel Identifier

Write a comprehension that takes a string as input and returns the vowels that appear in that string. Each vowel should only appear once, regardless of how many times it appears in the string.

_Te discretion is yours if you wish to include 'y'_


```python
text = "PyTexas is Awesome!"
vowels = {char for char in text if char in 'aeiouy'}
print(vowels)
```

# Exercise #3 - Create All Possible DND Classes/Races

Write a comprehension that takes two lists, one of various classes in DND and the other of various races in DND and generates every possible outcome.

```python
races = ["Human", "Elf", "Dragonborn", "Orc"]
classes = ["Wizard", "Barbarian", "Druid", "Rogue", "Paladin"]

all_options = [f"{race} {clas}" for race in races for clas in classes]

print(all_options)
```
